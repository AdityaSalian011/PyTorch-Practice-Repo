### Merging both dataframes

```python
final_df = pd.merge(df, df_country, how='left', on='Country Code')
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names = final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
plt.pie(x=country_values[:3], labels=country_names[:3], autopct='%.1f%%')
plt.show()
```

On the basis of piechart,
we can say that majority users are from India 
while only about 5% of users are from foreign countries

### Grouping columns

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings.head()
```

```python
ratings.rename(columns={0:'Rating freq'}, inplace=True)
```

```python
ratings.columns
```

```python
sns.barplot(x='Aggregate rating', y='Rating freq', data=ratings)
```

Observation
- Most ratings range between 3.0 to 3.8

#### Plotting based on color

```python
sns.barplot(x='Aggregate rating', y='Rating freq', data=ratings, hue='Rating color')
```

#### Frequency of each rating color

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','purple'])
```

#### Countries that have given 0 rating

```python
final_df[final_df['Rating color']=='White'].groupby(['Country']).size().reset_index()
```

```python
final_df[['Country', 'Currency']].groupby(['Country','Currency']).size().reset_index()
```
