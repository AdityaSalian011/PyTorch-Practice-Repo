## Feature Engineering

```python
df_train.info()
```

### Converting object values into numeric

```python
df_train['Age'].unique()
```

#### 1. Using pandas built in function .map()

```python
df_train['Age'] = df_train['Age'].map({
    '0-17': 0, 
    '18-25': 1, 
    '26-35': 2, 
    '36-45': 3, 
    '46-50': 4, 
    '51-55': 5, 
    '55+': 6
})
df_train.head()
```

```python
df_train['Age'].unique()
```

#### 2. Using sklearn's LabelEncoder 

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df_train['Age'])
df_train['Age'] = le.transform(df_train['Age'])
```

```python
categorical_features = [feature for feature in df_train.columns if df_train[feature].dtype == 'O']
categorical_features
```

```python
for categorical_feature in categorical_features:
    le = LabelEncoder()
    le.fit(df_train[categorical_feature])
    df_train[categorical_feature] = le.transform(df_train[categorical_feature])
```
