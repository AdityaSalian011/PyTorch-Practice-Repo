```python
import torch
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])

print(tensor_a)
print(tensor_b)
```

### 1. Addition

```python
tensor_a + tensor_b
```

```python
torch.add(tensor_a, tensor_b)
```

### 2. Subtraction

```python
tensor_b - tensor_a
```

```python
torch.sub(tensor_b, tensor_a)
```

### 3. Multiplication

```python
tensor_a * tensor_b
```

```python
torch.mul(tensor_a, tensor_b)
```

### 4. Division

```python
tensor_a/ tensor_b
```

```python
tensor_div = torch.div(tensor_a, tensor_b)
tensor_div
```

- Returns floating point value

```python
tensor_div.dtype
```

### 5. Remainder/ Modulus

```python
tensor_b % tensor_a
```

```python
torch.remainder(tensor_b, tensor_a)
```

### 6. Power

```python
tensor_a ** tensor_b
```

```python
torch.pow(tensor_a, tensor_b)
```

### Difference between tensor.add() and tensor.add_()

```python
print(tensor_a)
print(tensor_b)
```

```python
tensor_a.add(tensor_b)
```

```python
tensor_a
```

```python
tensor_a.add_(tensor_b)
```

```python
tensor_a
```

Key difference
- tensor.add_() reassigns the value of the variable, as opposed to tensor.add()

```python
tensor_a.sub_(tensor_b)
```

```python
tensor_a.mul_(tensor_b)
```

```python
tensor_a = tensor_a.float()  ## will give a run time error if int is not converted to float
tensor_a.div_(tensor_b)
```

```python
tensor_a.dtype
```

```python
tensor_a.mul_(tensor_b)
tensor_a.add_(tensor_b)
tensor_a.remainder_(tensor_b)
```

```python
tensor_a
```

```python
tensor_a.add_(tensor_b)
```

```python
tensor_a.pow_(tensor_b)
```
