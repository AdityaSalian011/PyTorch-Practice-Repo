```python
import torch
```

## 1. Reshape

```python
my_tensor = torch.arange(10)
my_tensor
```

```python
my_tensor2 = my_tensor.reshape(2,5)
my_tensor2
```

## 2. View

```python
my_tensor3 = torch.arange(10)
my_tensor3
```

```python
my_tensor4 = my_tensor3.view(5,2)
my_tensor4
```

#### We can use -1, if we dont't know the no. of items

```python
my_tensor5 = torch.arange(15)
my_tensor5
```

```python
my_tensor6 = my_tensor5.reshape(-1, 5)
my_tensor6
```

```python
my_tensor7 = my_tensor5.view(5, -1)
my_tensor7
```

#### Reshaped tensors share the same space in memory
- For both .reshape() and .view()

```python
my_tensor8 = torch.arange(10)
my_tensor8
```

```python
my_tensor9 = my_tensor8.reshape(5,2)
my_tensor9
```

```python
my_tensor8[1] = 4141
my_tensor8
```

```python
my_tensor9
```

#### Slicing in tensor

```python
my_tensor10 = torch.arange(10)
my_tensor10
```

```python
my_tensor10[7]
```

#### Selecting only column value

```python
my_tensor11 = my_tensor10.view(5,2)
my_tensor11
```

```python
my_tensor11[:, 1]
```

```python
my_tensor11[:, 1:]
```

This slices all the column from index 1 onward, keeping the column dimension intact
- Use tensor[:, 1] to get a flattened tensor
- Use tensor[:, 1:] if you want to preserve column dimensionality
